```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Housing.csv")
df.head()
```

```python
df.info()
```

```python
#Checked missing values. Dataset contains no null values.
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
#Created price_band column to categorize houses into Low, Medium, and High price ranges.
df["price_band"] = pd.qcut(df["price"], 3, labels=["Low", "Medium", "High"])
df.head()
```

```python
df.to_csv("cleaned_data.csv", index=False)
```

- Verified missing values
- Removed duplicate records
